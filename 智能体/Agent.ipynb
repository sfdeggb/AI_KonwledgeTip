## Agent 原理和实操

```python
Agent概念

Agent是以LLM为核心的能够自主分析，规划，并采用工具进行行动，并且能对任务完成结果进行反思的解决复杂任务的AI系统。
其中：
分析：LLM+提示词
规划：LLM+提示词
行动：LLM+tools(python函数/第三方MCP包装的服务)
反思：LLM+提示词

所以说目前Agent本质上还是提示词工程
```

```python
Agent范式
1. ReAct（Reasoning and Action）：是指Agent根据环境变化来做出自己的决策和行为。
2. Planing & Execute 事前计划好(当然可以再根据反馈再迭代)
```

```python
多Agent的协作范式
1.  two agents：类似于两个人协作完成任务,这个是最简单的协作模式
2.  顺序agents：顺序Agents指的是一个任务过来，agentA去轮询其它agents，每次都带着问题和前几个人的沟通结果，最终得到解决方案.
3.  Group agents：一个任务过来，给到manager，manager负责分发任务给不同的agent，获得反馈，广播，并选择下一个agent
4.  Nested agents：
```
